### Usability of Mobilenet from TensorFlow for image classification.

MobileNets were made possible with the hard work of many engineers and researchers throughout Google.

MobileNets are small, low-latency, low-power models parameterized to meet the resource constraints of a variety of use cases. They can be built upon for classification, detection, embeddings and segmentation similar to how other popular large scale models, such as Inception, are used.

https://github.com/keras-team/keras/blob/v2.7.0/keras/applications/mobilenet.py#L80-L311

```python
'''Importing libraries'''

import keras
from keras import backend as K
from keras.layers.core import Dense, Activation
from tensorflow.keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
from keras.models import Model
from keras.applications import imagenet_utils
from keras.layers import Dense,GlobalAveragePooling2D
from keras.applications.mobilenet import MobileNet
from keras.applications.mobilenet import preprocess_input
import numpy as np
from IPython.display import Image
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Creating a function that shearch in a existe folder related images.

def prepare_image(file):
    img = image.load_img(file, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array_expanded_dims = np.expand_dims(img_array, axis=0) # (1, 224, 224)
    return keras.applications.mobilenet.preprocess_input(img_array_expanded_dims)

Image(filename='/content/drive/MyDrive/file/German_Shepherd.jpg')
```

# Load mobilenet model

```python
""" Loading the model to perform the preditions inferences"""

mobile = keras.applications.mobilenet.MobileNet()
```

# Inference

```python
preprocessed_image = prepare_image('/content/drive/MyDrive/file/German_Shepherd.jpg')
predictions = mobile.predict(preprocessed_image)
results = imagenet_utils.decode_predictions(predictions)
results
```
