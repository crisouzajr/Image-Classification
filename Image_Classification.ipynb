From a trained image classification model created on the web site www.teachablemachine.com, is possible to define a set images.

Example? If we use 3 label were each one indicate a "raça" of dogs, the model is able to define exactly each one.

```python
# Import libraries

from tensorflow.keras.models import load_model
from PIL import Image, ImageOps
import numpy as np
```

```python
np.set_printoptions(suppress=True)
```

```python
model = load_model('keras_model.h5')
```

```python
data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)
```

```python
# ler a imagem (escolher imagem aleatória e colocar ao lado)

image = Image.open('2203.png')
image
```

```python
# redimensionar a imagem
size = (224, 224)
image = ImageOps.fit(image, size, Image.ANTIALIAS)
image
```

```python
image_array = np.asarray(image)
```

```python
image.show()
```

```python
# normalizing pixels

norm_array = (image_array.astype(np.float32) /127)-1
norm_array.shape
```

```python
# dimensão que seria o número de imagens

data[0] = norm_array 
data
```

```python
prediction = model.predict(data)
print(prediction)
```
